```python
import turtle as dt
pen= dt.Turtle()
#here we are creating a function or method for our face features.    
def eyes(color, radius):
    pen.down()
    pen.fillcolor(color)
    pen.begin_fill()
    pen.circle(radius)
    pen.end_fill()
    pen.up()

def nose(color, radius):
    pen.down()
    pen.fillcolor(color)
    pen.begin_fill()
    pen.circle(radius)
    pen.end_fill()
    pen.up()

#here we start drawing the face 

pen.fillcolor('orange')
pen.begin_fill()
pen.circle(100)
pen.end_fill()
pen.up()

#Now we are drawing the eyes and giving them properties
pen.goto(-40, 120)#here are Horizontal and vertical alignment features
eyes('white', 15)
pen.goto(-37, 125) #Here is our left eye
eyes('black', 5)

pen.goto(40, 120)
eyes('white', 15) #Here its our right eye
pen.goto(40, 125)
eyes('black', 5)

#Here we draw the nose
pen.goto(0, 75)
nose('black', 8)
#here we draw the ears
pen.goto(-100, 100)
nose('black', 4)
pen.goto(100, 100)
nose('black', 4)
#here we draw mouth 
pen.goto(-40, 85)
pen.down()
pen.right(90)
pen.circle(40, 180)
pen.up()

#Tongue
pen.goto(-10, 45)
pen.down()
pen.right(180)
pen.fillcolor('red')
pen.begin_fill()
pen.circle(10, 180)
pen.end_fill()
pen.hideturtle()

         
```
